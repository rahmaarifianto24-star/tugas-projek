```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow
import tensorflow as tf
```

```python
y_tf2 = tf.Variable(5.78)
```

```python
x_tf.shape
```

```python
y_tf = tf.Variable(3, dtype=tf.int16)
```

```python
x_tf + y_tf2
```

```python
tf_sum = tf.add(x_tf, y_tf2)
tf_sum
```

```python
tf_sum.numpy()
```
